### **SISTEMA DE ECUACIONES LINEALES**

###Resolución de ecuaciones lineales mediante el uso de matrices

***Matrices***

Se define matrices a un conjunto ordenado de datos que se representan por filas y columnas, también es utilizado para resolver sistemas de ecuaciones lineales donde se tiene un resultado constante.

La notación matricial se obtiene de la siguiente manera:

$$A_x=b$$
Donde *A* :es la matriz de coeficientes [n x n]

*b* : es el vector de terminos independientes

*x* : es es vector de incognitas


*Generamos una matriz de 10 x 10*

```python

import numpy as np  ## Inicialmente importo la libreria numpy encargada de manipular estructuras numericas y la nombro con el alias de np
from tabulate import tabulate ## importo libreria para impresion de tablas por consola
from termcolor import colored ## importo libreria de colores para textos
import matplotlib.pyplot as plt 
from numpy import arange

n = 10 # declaro la variable n con un valor inicial de 10
A = np.random.rand(n,n) #declaro la variable (A) de tipo float en un array con longitud maxima de 10 numeros aleatorios 0-1  x 10 vectores de datos maximos 
b = np.random.rand(n) #declaro la variables (b) de tipo float en un array  con longitud maxima de 10 numeros aleatorios de 0-1

#ciclo for para los headers A tablas
titulos = ["VECTOR"]
for indice in range(0, 10):
    titulos.append("VALOR %d" %(indice))    

#Imprimo los valores de la matriz A
print(colored("\n VALORES MATRIZ A","green"))
print(tabulate(A, headers=titulos, showindex="always", tablefmt="presto"))

#Imprimo los valroes del vector b
print(colored("\n VALORES VECTOR b","green"))
print(tabulate([b], headers=titulos, showindex="always", tablefmt="presto"))

alpha = 3 * n ## declaro la varialbe alpha multiplicando 3 * el valor de la variable n(10) para un resultado de 30

print("\n")
print(colored("VALOR VARIABLE alpha = " + str(alpha),"green"))

#realizo el producto de las matrices aplicando una transposicion de los resultados para el tiempo de ejecucion debido a que
#las matrices de datos no son iguales; al producto de la matriz obtenida le sumo el valor de alpha (30) multiplicado por una matrix de 10x10
#generada por la funcion de matriz de identidad.

ADD = np.dot(A,A.T)+ alpha * np.identity(n)

A  = ADD  #actualizo los valores de la nueva matriz generada

#Imprimo los valores de mi matriz ADD

print(colored("\nVALORES MATRIZ ADD", 'green'))
print(tabulate(ADD, headers=titulos, showindex="always", tablefmt="presto"))
```

###Evaluo si la diagonal dominante

Es dominante si a cada fila de la matriz, la magnitud de la entrada diagonal en una fila es mayor o igual que la suma de las magnitudes de todas las demás (no diagonales) entradas en esa fila. 

$$\left|a_{ii}\right|\geq\sum_{j\ \neq\ i}\left|a_{ii}\right|$$

*i* es la entrada de la fila

*j* es la entrada de la columna

## Metodo Jacobi

Este método permite hallar las aproximaciones a una solución de sistemas de ecuaciones lineales, utilizando los valores iniciales para la primera aproximación, luego los de la primera para la segunda y así sucesivamente; en este método el cálculo de cada variable es independiente por lo tanto ninguna variable depende de la otra. Este método no es tan eficiente como el de Gauss Seidel.

**Formula**
$$x_i^{(k+1)}=\frac{1}{a_{ii}}(b_i-\sum_{j \neq i}{a_{ij}x_j^k}),i\ =\ 1,2,3...$$


```python
#####################################
#Jacobi
#####################################
print("\n")
toleranciaAproximacion = 1e-6
errorJacobi = 100000
iteracionesMaximas = 100
n = len(A)
x = np.zeros(n)
xl= np.zeros(n)
numeroIteracion = 0
iteracionesArray = []
Error= []

while numeroIteracion < iteracionesMaximas:
    for i in range(n):
        temp = b[i]
        for j in range(n):
            if i != j :
                temp = temp - A[i][j] * x[j]
        x[i] = temp / A[i][i]
  
    errorJacobi = np.linalg.norm(x-xl)
    xl = np.copy(x)

    if (errorJacobi > 0):      
      print(colored("Aproximación al resultado en la iteracion %d ","yellow") %numeroIteracion)
      print(colored("Con los valores: ","blue"))
      print(tabulate([x], headers=titulos, showindex="always", tablefmt="simple"))
      print(colored("Su margen de error es de ","blue"), errorJacobi)
      print("\n")
    else:
      print(colored("Aproximación al resultado en la iteracion %d ","yellow") %numeroIteracion)
      print(colored("Con los valores: ","red"))
      print(tabulate([x], headers=titulos, showindex="always", tablefmt="simple"))
      print(colored("Su margen de error es de ","red"), errorJacobi)

    numeroIteracion += 1
    iteracionesArray.append(numeroIteracion)
    Error.append(errorJacobi)
    if (errorJacobi < toleranciaAproximacion):      
      print(colored("Valor aproximado a la solucion se encuentra en la iteracion" ,"green"),numeroIteracion)
      print(colored("Con los valores: ","green"),x)
      print(colored("Su margen de error es de ","green"), errorJacobi)
      print("\n")
      break
```
