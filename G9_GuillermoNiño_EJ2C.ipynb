## Regresión Lineal 

La regresión se refiere a la cantidad de cambio que experimenta una varible (y) dependiente con respecto a una variable independiente (x) mediante una función matemática (recta, parabola, etc.) a partir de un conjunto de datos, este concepto está relacionado con la correlación, la regresoón estudía la naturaleza de la relación entre dos variables dependientes mientras la correlación estudia la estrechez de la relación entre esas dos variables una dependiente de la otra.

*Regresión lineal y ajuste por mínimos cuadrados*

La aproximación por mínimos cuadrados es ajustar una linea recta a un conjunto de observaciones definidas por puntos: ($x_{1}$,$y_{1}$),($x_{2}$,$y_{2}$)...,($x_{n}$,$y_{n}$).

La expresión matemática para la recta es:
$$y=a_0+a_1x+e$$

Siendo **$a_0 , a_1$** coeficientes que representan la intersección con el eje y la pendiente respectivamente.

**$e$** es el error entre la recta y los puntos

**Hallar ajuste de una linea recta**

*Formula para hallar la pendiente.*
$$a_1=\ \frac{n\sum(x_iy_i)-\sum(x_i)\sum(y_i)}{n\sum(x_i^2)-{(\sum{x_i)}}^2}$$

*Formula para hallar el intercepto.*
$$a_0=\bar{y}-a_1\bar{x}$$
Donde $\bar{y},\bar{x}$ son las medias de *y* y *x*, respectivamente.

**Ajuste de error**

El residuo en la regresión lineal representa la distancia vertical entre un dato y la linea recta.
Para obtener un error mínimo se hace una *suma de errores residuales*:
$$Sr=\sum_{i=1}^{n}e_i^2\ =\ \sum_{i=1}^{n}{(y_1-a_0-a_1x_1)}^2$$

*Estimación del error*

$$S_{y/x}=\sqrt{\frac{S_r}{n-2}}$$

**Coeficiente de correlación**

Es la medida específica que cuantifica la intensidad de la relación lineal entre dos variables en un análisis de correlación.

$$r^2=\sqrt{\frac{S_r}{n-2}}$$

## Punto 2 Ejercicio C

Se realizó un experimento con un fluido al que se le transfirió calor de forma constante. La temperatura se tomó cada dos minutos durante los  primeros 14 minutos. La siguiente tabla resume los resultados obtenidos, y se observa que la temperatura aumenta de forma lineal. 

tiempo (minutos) | temperatura (°C)
:--------: | -------
0  | 25
2  | 28,1
4  | 35,2
6  | 36,2
8  | 41,3
10 | 42,1
12 | 47,9
14 | 54,2

Realice un programa en Python empleando Google Colab que:

- Ajuste los datos a una línea recta empleando el método de los mínimos cuadrados, realizando una descripción teórica detallada.
- Calcule la pendiente y la intersección, así como el error estándar y el coeficiente de correlación.
- Realice una gráfica en donde se muestren los datos originales y la recta que ajusta.
- Haga un análisis de resultados basándose en la teoría y en los resultados obtenidos. Debe incluir referencias a la bibliografía empleada usando las normas APA. Comentarios sin referencias no serán tenidos en cuenta.
- Para cada método haga una descripción teórica (las ecuaciones deberán estar escritas en LaTeX) incluyendo referencias bibliográficas empleadas siguiendo las normas APA.

```python
import numpy as np
import matplotlib.pyplot as plt

# Variables

x = np.array([0,2,4,6,8,10,12,14])  #tiempo
y =np.array ([25, 28.1, 35.2, 36.2, 41.3, 42.1, 47.9, 54.2]) # temperatura

n = len (x)
sumx = sum(x) 
sumy = sum(y)
sumx2 = sum((x)**2)
sumxy = sum (x*y)
promx = sum (x) / n
promy = sum (y) / n
st = 0
sr = 0
a1 = (n * sumxy - sumx * sumy) / (n * sumx2 - sumx*sumx) #pendiente
a0 = promy - a1*promx #intersección

for i in range (n):
    st = st + (y[i] - promy)**2
    sr = sr + (y[i] - a1 * x[i] - a0)**2

syx = np.sqrt(sr/(n - 2)) #error estandar
R2 = (st-sr)/ st # coeficiente de correlación; es mejor si se acerca a 1

print("La pendiente es: ", a1)
print("La intersección es: ", a0)
print("El error estandar es: ", syx)
print ("El coeficiente de correlación es : ", R2)

t = np.poly1d ([a1,a0])
print (t)

plt.plot(x, y, 'o', label='variable') #gráfica en forma de puntos
plt.plot(x, a1*x+a0 , label='Recta ajustada') 
plt.xlabel('tiempo') 
plt.ylabel('temperatura')
plt.title('Regresión lineal')
plt.grid() #rejillas
plt.legend() #muestra los labels
plt.show() #imprime la gráfica
```
