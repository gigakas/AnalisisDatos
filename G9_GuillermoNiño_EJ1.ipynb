### **SISTEMA DE ECUACIONES LINEALES**

Son un conjunto de funciones matemáticas que al obtener los resultados mantienen una linealidad al representarlos en un plano cartesiano en sus ejes *x* y *y*.

###Resolución de ecuaciones lineales mediante el uso de matrices

***Matrices***

Se define matrices a un conjunto ordenado de datos que se representan por filas y columnas, también es utilizado para resolver sistemas de ecuaciones lineales donde se tiene un resultado constante.

La notación matricial se obtiene de la siguiente manera:

$$A_x=b$$
Donde *A* :es la matriz de coeficientes [n x n]

*b* : es el vector de terminos independientes

*x* : es es vector de incognitas


*Generamos una matriz de 10 x 10*

```python

import numpy as np  ## Inicialmente importo la libreria numpy encargada de manipular estructuras numericas y la nombro con el alias de np
from tabulate import tabulate ## importo libreria para impresion de tablas por consola
from termcolor import colored ## importo libreria de colores para textos
import matplotlib.pyplot as plt 
from decimal import Decimal

n = 10 # declaro la variable n con un valor inicial de 10
A = np.random.rand(n,n) #declaro la variable (A) de tipo float en un array con longitud maxima de 10 numeros aleatorios 0-1  x 10 vectores de datos maximos 
b = np.random.rand(n) #declaro la variables (b) de tipo float en un array  con longitud maxima de 10 numeros aleatorios de 0-1

#ciclo for para los headers A tablas
titulos = ["VECTOR"]
for indice in range(0, 10):
    titulos.append("VALOR %d" %(indice))    

#Imprimo los valores de la matriz A
print(colored("\n VALORES MATRIZ A","green"))
print(tabulate(A, headers=titulos, showindex="always", tablefmt="presto"))

#Imprimo los valroes del vector b
print(colored("\n VALORES VECTOR b","green"))
print(tabulate([b], headers=titulos, showindex="always", tablefmt="presto"))

alpha = 3 * n ## declaro la varialbe alpha multiplicando 3 * el valor de la variable n(10) para un resultado de 30

print("\n")
print(colored("VALOR VARIABLE alpha = " + str(alpha),"green"))

#realizo el producto de las matrices aplicando una transposicion de los resultados para el tiempo de ejecucion debido a que
#las matrices de datos no son iguales; al producto de la matriz obtenida le sumo el valor de alpha (30) multiplicado por una matrix de 10x10
#generada por la funcion de matriz de identidad.

ADD = np.dot(A,A.T)+ alpha * np.identity(n)

A = ADD #actualizo el valor de la matriz A con los datos de ADD

#Imprimo los valores de mi matriz ADD

print(colored("\nVALORES MATRIZ ADD", 'green'))
print(tabulate(ADD, headers=titulos, showindex="always", tablefmt="presto"))
```

<h1>Se determina si la matriz es singular o no singular</h1>

Para evaluar si es singular (que tiene una única solución) la determinante de la matriz obtenida debería ser  igual a cero, de lo contrario da un número mayor a cero.

$$A_N=\ \left[\begin{matrix}a_{1,1}&\begin{matrix}a_{1,2}&a_{1,N-1}&a_{1,N}\\\end{matrix}\\\begin{matrix}a_{2,1}\\a_{N-1,1}\\a_{N,1}\\\end{matrix}&\begin{matrix}a_{2,2}&a_{2,N-1}&a_{2,N}\\a_{N-1,2}&a_{N-1,N-1}&a_{N-1,N}\\a_{N,2}&a_{N-N-1}&a_{N-N}\\\end{matrix}\\\end{matrix}\right]$$


```python
# Evaluo si el determinante de mi matriz  es igual a cero 
# determ = np.linalg.det(A)
# print(colored("VALOR DE LA DETERMINANTE ES = " + str(determ),"green"))

#Defino una funcion determinante con los parametros matriz que sera mi matriz de datos y multiplicador 
def determinante(matriz, multiplicador):
    tamanio = len(matriz)
    if tamanio == 1:
        return multiplicador * matriz[0][0]
    else:
        signo = -1
        respuesta = 0
        for i in range(tamanio):
            m = []
            for j in range(1, tamanio):
                buffer = []
                for k in range(tamanio):
                    if k != i:
                        buffer.append(matriz[j][k])
                m.append(buffer)
            signo *= -1
            respuesta = respuesta + multiplicador * determinante(m, signo * matriz[0][i])
    return respuesta #retorno el valor del calculo realizado en las iteraciones

#procedo a llamar la funcion y paso los valores de la matriz A y un factor de multiplicacion de 1 con  resultado de coma flotante .6
determinante = determinante(A, 1)
if(determinante > 0):
    print(colored("VALOR DE LA DETERMINANTE ES = %s - por lo cual es una matriz no singular","green") %determinante)    
else:
    print(colored("VALOR DE LA DETERMINANTE ES = %s - por lo cual es una matriz singular (unica solucion)","red") %determinante)    
```

<h1>Evaluo si la diagonal dominante</h1>

Es dominante si a cada fila de la matriz, la magnitud de la entrada diagonal en una fila es mayor o igual que la suma de las magnitudes de todas las demás (no diagonales) entradas en esa fila. 

$$\left|a_{ii}\right|\geq\sum_{j\ \neq\ i}\left|a_{ii}\right|$$

*i* es la entrada de la fila

*j* es la entrada de la columna

```python
#####################################
#matriz diagonal dominante evaluacion
#####################################
numItems = len(b)
def matrizDominante(matriz,numItems):
      for i in range(0,numItems):
          sumaVector = 0
          for j in range(0,numItems):
              sumaVector = sumaVector + abs(matriz[i][j])

          sumaVector = sumaVector - abs(matriz[i][i])

          if (abs(matriz[i][i])< sumaVector):
             print(colored("La matriz no es diagonal dominante en el vector %d","red") %i)
          else:
            print(colored("La matriz  es diagonal dominante en el vector %d","green") %i)

matrizDominante(A,numItems)
```

<h1>Metodo de Gauss Seidel</h1>
Es una versión mejorada del metodo Jacobi también conocido como método de desplazamiento sucesivo, el cúal permite mejorar el rendimiento del calculo de matrices de gran tamaño simplificando los procesos de calculo por iteraciones con el fin de llegar a la tolerancia deseada.

*Determinantes*

$$\left[A\right]\left\{X\right\}=\left\{B\right\}$$

[A] Es la matriz de coeficientes

*Formula*
$$x_i^{(k+1)}=\frac{-\sum_{1 \le j \le i-1}{a_{ij}x_j^k+b_i}}{a_{ii}},i\ =\ 1,...,n\ ( ^*)$$

```python
#####################################
#gauss seidel
#####################################

salida= np.zeros(len(A))
xlgauss= np.zeros(len(A))
toleranciaAproximacion = 1e-6
iteracionesMaximas = 40
resultado = 0
iteracionGaussSediel = []
errorGauseSediel = 0
erroresGausSedielArray = []

#creo una funcion que recibe tres argumentos matriz, salida (valores a calcular) e igualdad (resultado obtenido de las operaciones de la ecuacion lineal)   
def gauss_seidel(matriz,salida,igualdad):
    #Calculo el tamaño de la matriz       
    n = len(matriz)      
    # for anidaddo de dos niveles para calcular los elementos de la matriz de 2x2
    for j in range(0, n):        
        # variable temporal para almacenar los valores de mi vector de datos
        temp = igualdad[j]                            
        # realizo 
        for i in range(0, n):     
            if(j != i):
                temp-=matriz[j][i] * salida[i]

        # actualizo el valor en mi vector de salida     
        salida[j] = temp / matriz[j][j]
    #returno el vector de datos con la salida de los calculos realizados               
    return salida
    
#aplico la funcion creada y almaceno los datos generados en arreglos para su posterior impresion y graficacion.     
for i in range(iteracionesMaximas):    
    resultado = gauss_seidel(A, salida, b)
    iteracionGaussSediel.append(i)

    errorGauseSediel = np.linalg.norm(salida-xlgauss)      
    xlgauss = np.copy(salida)  
  
    if (errorGauseSediel < 0):      
      print(colored("Aproximación al resultado en la iteracion %d ","yellow") %i)
      print(colored("Con los valores: ","green"))
      print(tabulate([resultado], headers=titulos, showindex="always", tablefmt="simple"))
      print(colored("Su margen de error es de ","blue"), errorGauseSediel)
      print("\n")
    else:
      print(colored("Aproximación al resultado en la iteracion %d ","yellow") %i)
      print(colored("Con los valores: ","green"))
      print(tabulate([resultado], headers=titulos, showindex="always", tablefmt="simple"))
      print(colored("Su margen de error es de ","red"), errorGauseSediel)
      print("\n")
        
    erroresGausSedielArray.append(errorGauseSediel)
    if (errorGauseSediel < toleranciaAproximacion):      
      print(colored("Valor aproximado a la solucion se encuentra en la iteracion %d" ,"green") %i)
      print(colored("Con los valores: ","green"),salida)
      print(colored("Su margen de error es de ","green"), errorGauseSediel)
      print("\n")
      break


plt.figure(figsize=(16,9)) # especifíca el tamaño de la imagen
plt.plot(iteracionGaussSediel,erroresGausSedielArray,"blue",Label="Gauss Seidel")
plt.scatter(iteracionGaussSediel,erroresGausSedielArray,color="blue")
plt.xlabel("Iteraciones")
plt.ylabel("Margen de error")
plt.title("Gráfico de relación iteraciones - aproximación de resultado metódo Gauss Seidel")
plt.grid()
plt.xlim([1, 30])
plt.xticks(range(0,iteracionesMaximas+1,2))
plt.legend()
plt.show()

```

## Metodo Jacobi

Este método permite hallar las aproximaciones a una solución de sistemas de ecuaciones lineales, utilizando los valores iniciales para la primera aproximación, luego los de la primera para la segunda y así sucesivamente; en este método el cálculo de cada variable es independiente por lo tanto ninguna variable depende de la otra. Este método no es tan eficiente como el de Gauss Seidel.

**Formula**
$$x_i^{(k+1)}=\frac{1}{a_{ii}}(b_i-\sum_{j \neq i}{a_{ij}x_j^k}),i\ =\ 1,2,3...$$

```python
#####################################
#Jacobi
#####################################
print("\n")
toleranciaAproximacion = 1e-6
errorJacobi = 100000
iteracionesMaximas = 40
n = len(A)
x = np.zeros(n)
xl= np.zeros(n)
numeroIteracion = 0
iteracionesArray = []
Error= []
#evaluo si el numero de iteraciones es menor que las iteraciones maximas para deterne el bucle while
while numeroIteracion < iteracionesMaximas:
    for i in range(n): # inicio mi iteracion con el tamaño dado por n  de nuestra matrix para las filas
        temp = b[i] #almaceno un valor termporal para mi primer dato a calcular este sera el que almacene el valor de los datos calculados
        for j in range(n): # #inicio la iteracion para el caso de la columnas tomando como valor n el limite de nuestra matriz
            if i != j :
                temp = temp - A[i][j] * x[j] #calculo el valor de temp restando a temp el valor de la matriz en poscion [i][j] y luego multplicandolo por nuestro vector de salida x  el cual almacena los datos calculados posteriormente.
        x[i] = temp / A[i][i] #divido el valor obtenido en temp por el valor de la matriz A en[i][i]
  
    errorJacobi = np.linalg.norm(x-xl) # procedo a restar nuestros vectores para alamcenar los valores de margen de error dados al realizar la operacion
    xl = np.copy(x) #copio los valores almacenados en cada iteracion en x en el array xl

    if (errorJacobi > 0):    #evaluo si el valor de error es superior y presento los datos almacenados en x en una tabla  
      print(colored("Aproximación al resultado en la iteracion %d ","yellow") %numeroIteracion)
      print(colored("Con los valores: ","blue"))
      print(tabulate([x], headers=titulos, showindex="always", tablefmt="simple"))
      print(colored("Su margen de error es de ","blue"), errorJacobi)
      print("\n")
    else: #si se sobrepasa de cero presento una alerta y su respectiva tabla de datos
      print(colored("Aproximación al resultado en la iteracion %d ","yellow") %numeroIteracion)
      print(colored("Con los valores: ","red"))
      print(tabulate([x], headers=titulos, showindex="always", tablefmt="simple"))
      print(colored("Su margen de error es de ","red"), errorJacobi)

    numeroIteracion += 1
    iteracionesArray.append(numeroIteracion)
    Error.append(errorJacobi)  #almaceno los datos de error en un array
    if (errorJacobi < toleranciaAproximacion): #si el valor de error es diferente al dado en la aproximacion salgo del ciclo while     
      print(colored("Valor aproximado a la solucion se encuentra en la iteracion" ,"green"),numeroIteracion)
      print(colored("Con los valores: ","green"),x)
      print(colored("Su margen de error es de ","green"), errorJacobi)
      print("\n")
      break

#impresion de los resultados en una grafica
plt.figure(figsize=(16,9)) # especifíca el tamaño de la imagen
plt.plot(iteracionesArray,Error,"red",Label="Jacobi")
plt.scatter(iteracionesArray,Error,color="red")
plt.xlabel("Iteraciones")
plt.ylabel("Margen de error")
plt.title("Gráfico de relación iteraciones - aproximación de resultado - metódo JACOBI")
plt.grid()
plt.xlim([1, 30])
plt.xticks(range(0,iteracionesMaximas+1,2))
plt.legend()
plt.show()
```

<h1>Procedimiento resolucion mediante S.O.R</h1>
El método de Gauss Seidel con relajación, procede de la misma manera que el anterior método de Gauss Seidel, pero basándose en una técnica que intenta mejorar la convergencia del método, mediante el refinamiento de la aproximación actual, para esto se utiliza un coeficiente de relajación (w).

El metodo es muy parecido al usado por Guass-Seidel, la unica diferencia es que se debe multiplicarl el valor de relajacion por cada uno de los valores obtenidos en cada iteracion.


*Formula*
$$x_i^{(k+1)}=\ w\ *\ x_i^{(k+1)}+(1-w)*x_i^k\ $$




```python
#####################################
#S.O.R
#####################################

salidaSOR= np.zeros(len(A))
xlSOR= np.zeros(len(A))
toleranciaAproximacion = 1e-6
iteracionesMaximas = 40
resultadoSOR = 0
iteracionesSOR = []
erroresSOR = 0
erroresSORarray = []
factorRelajacion = 1.2  #factor de relajacion comprendido entre 0 y 2 


#creo una funcion que recibe tres argumentos matriz (vector de 2x2 con los datos a evaluar), salida (valores a calcular) e igualdad (resultado obtenido de las operaciones de la ecuacion lineal)   
def SOR(matriz,salida,igualdad):
    #Calculo el tamaño de la matriz       
    n = len(matriz)      
    # for anidaddo de dos niveles para calcular los elementos de la matriz de 2x2
    for j in range(0, n):        
        # variable temporal para almacenar los valores de mi vector de datos
        temp = igualdad[j]                            
        # realizo el calculo de los valores para cada iteracion
        for i in range(0, n):     
            if(j != i):
                temp-= matriz[j][i] * salidaSOR[i]

        # actualizo el valor en mi vector de salida     
        salidaSOR[j] = temp / matriz[j][j]
    #retorno el vector de datos con la salida de los calculos realizados               
    return salida
    
#aplico la funcion creada y almaceno los datos generados en arreglos para su posterior impresion y graficacion.     
for i in range(iteracionesMaximas):    
    resultadoSOR = SOR(A, salidaSOR, b)
    iteracionesSOR.append(i)
    erroresSOR = np.linalg.norm((factorRelajacion *salidaSOR) - (factorRelajacion * xlSOR))
       
    xlSOR = np.copy(salidaSOR)  
  
    if (erroresSOR < 0):  #evaluo si el valor de error es menor a 0 
      print(colored("Aproximación al resultado en la iteracion %d ","yellow") %i)
      print(colored("Con los valores: ","green"))
      print(tabulate([resultado], headers=titulos, showindex="always", tablefmt="simple"))
      print(colored("Su margen de error es de ","blue"), erroresSOR)
      print("\n")
    else:  #si se sobrepasa de cero presento una alerta y su respectiva tabla de datos
      print(colored("Aproximación al resultado en la iteracion %d ","yellow") %i)
      print(colored("Con los valores: ","green"))
      print(tabulate([resultado], headers=titulos, showindex="always", tablefmt="simple"))
      print(colored("Su margen de error es de ","red"), erroresSOR)
      print("\n")
        
    erroresSORarray.append(erroresSOR) #almaceno los datos de error en un array
    if (erroresSOR < toleranciaAproximacion):  #si el valor de error es diferente al dado en la aproximacion salgo del ciclo for     
      print(colored("Valor aproximado a la solucion se encuentra en la iteracion %d" ,"green") %i)
      print(colored("Con los valores: ","green"),x)
      print(colored("Su margen de error es de ","green"), erroresSOR)
      print("\n")
      break
      



plt.figure(figsize=(16,9)) # especifíca el tamaño de la imagen
plt.plot(iteracionesSOR,erroresSORarray,"green",Label="S.O.R")
plt.scatter(iteracionesSOR,erroresSORarray,color='green')
plt.xlabel("Iteraciones")
plt.ylabel("Margen de error")
plt.title("Gráfico de relación iteraciones - aproximación de resultado metódo JACOBI")
plt.grid()
plt.xlim([1, 30])
plt.xticks(range(0,iteracionesMaximas+1,2))
plt.legend()
plt.show()      
```
